```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib 
import seaborn as sns

%matplotlib inline
```

```python
# Read CSV files containing sales data into pandas DataFrame
temperature_df = pd.read_csv('/Users/yejinchoi/MenuOptimization/temperature_by_season.csv')
black_tea_category_df = pd.read_csv('/Users/yejinchoi/MenuOptimization/black_tea_by_category.csv')
position_df = pd.read_csv('/Users/yejinchoi/MenuOptimization/position_sales.csv')
```

---

### 1. Analyze Sales by Season and Beverage Temperature

```python
temperature_df
```

```python
# Create a grouped bar graph to visualize sales by temperature and season
temperature_df.plot(x='season',
                    kind='bar',
                    stacked=False,                                        # Plot separate bar for each temperature group
                    title='Sales of Hot and Iced Products by Category',   # Set title
                    color=('tab:orange','tab:blue'),                      # Set colors
                    xlabel='Season',                                      # Set x-axis label
                    ylabel='Total Sales (Normalized)'                     # Set y-axis label
                    )

# Tilt the x-axis labels
plt.xticks(rotation=0)

# Display the graph
plt.show()
```

---

### 2. Analyze Sales by Whether the Product is Black Tea-Related or Not

```python
black_tea_category_df
```

```python
# Rename columns
black_tea_category_df.rename(columns={'black_tea_sales': 'Black Tea Products', 'non_black_tea_sales': 'Non-Black Tea Products'}, inplace=True)

# Create a grouped bar graph to visualize sales by temperature and season
black_tea_category_df.plot(x='category',
                           kind='bar',
                           stacked=False,                               # Plot separate bar for each product group
                           title='Average Sales of Black Tea and Non-Black Tea Products by Category',   # Set title
                           color=('tan','gray'),                        # Set colors
                           xlabel='Category',                           # Set x-axis label
                           ylabel='Average Sales (Normalized)'          # Set y-axis label
                           )

# Tilt the x-axis labels
plt.xticks(rotation=0)

# Display the graph
plt.show()
```

---

### 3. Analyze Sales by the Product's Position in the Menu

```python
position_df
```

```python
# Rename columns
position_df.rename(columns={'target_product_sales': 'Target Products', 'other_product_sales': 'Other Products'}, inplace=True)

# Create a grouped bar graph to visualize sales by position and category
position_df.plot(x='category',
                kind='bar',
                stacked=False,                                                # Plot separate bar for each product group
                title='Average Sales of Products by Position and Category',   # Set title
                color=('tab:red','tab:green'),                                # Set colors
                xlabel='Category',                                            # Set x-axis label
                ylabel='Average Sales (Normalized)'                           # Set y-axis label
                )

# Tilt the x-axis labels
plt.xticks(rotation=0)

# Display the graph
plt.show()
```
